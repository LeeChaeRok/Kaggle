### 이상치 처리

```python
sns.boxplot(c_data["DURATION"])
```

```python
y= c_data[(c_data["DURATION"] > 800)].index
c_data.drop(y,axis=0,inplace=True)
```

```python
def classify(x,column):
    q1_D= x[column].quantile(0.25) 
    q2_D= x[column].quantile(0.5) 
    q3_D= x[column].quantile(0.75) 

    c_data.loc[ x[column] <= q1_D, column] = 0
    c_data.loc[ (x[column] <= q2_D) & (x[column] > q1_D), column] = 1
    c_data.loc[ (x[column] <= q3_D) & (x[column] > q2_D), column] = 2
    c_data.loc[ x[column] > q3_D, column] = 3
    return x
```

```python
classify(c_data,'DURATION')
```

```python
c_data["DURATION"].value_counts()
```

```python
sns.boxplot(c_data["AMOUNT"])
```

```python
c_data.loc[ c_data['AMOUNT'] > 30000, 'AMOUNT'] = 30000
```

```python
classify(c_data,'AMOUNT')
```

```python
def classify_A(x,column):

    c_data.loc[ x[column] <= 20, column] = 0
    c_data.loc[ (x[column] <= 40) & (x[column] > 20), column] = 1
    c_data.loc[ (x[column] <= 60) & (x[column] > 40), column] = 2
    c_data.loc[ x[column] > 60, column] = 3
    return x
```

```python
classify_A(c_data,'AGE')
```

### 변수 선정 by t-test

```python
import scipy.stats as stats
from scipy.stats import f
from scipy.stats import ttest_ind
table = pd.DataFrame(data={'Variable':[], 'delete':[],'equal/not equal':[], 'p-value for f-test':[], 'p-value for t-test':[]})

for i in range(1, 41):
    lresult = stats.levene(c_data.iloc[:,0], c_data.iloc[:,i])
    f_value = pd.DataFrame(lresult)
    if f_value.iloc[1,0] > 0.05:
        abc = pd.DataFrame(ttest_ind(c_data.iloc[:,0], c_data.iloc[:,i], equal_var = True))
        if abc.iloc[1,0] > 0.05:
            table.loc[i] = [c_data.iloc[:,i].name, 'delete', 'equal', f_value.iloc[1,0], abc.iloc[1,0]]
        else:
            table.loc[i] = [c_data.iloc[:,i].name, 'not delete', 'equal', f_value.iloc[1,0], abc.iloc[1,0]]
    else:
        cba = pd.DataFrame(ttest_ind(c_data.iloc[:,0], c_data.iloc[:,i], equal_var = False))
        if cba.iloc[1,0] > 0.05:
            table.loc[i] = [c_data.iloc[:,i].name, 'delete', 'not equal', f_value.iloc[1,0], cba.iloc[1,0]]
        else:
            table.loc[i] = [c_data.iloc[:,i].name, 'not delete', 'not equal', f_value.iloc[1,0], cba.iloc[1,0]]
table
```

## 분석 by Cross Validation

```python
dt_clf = DecisionTreeClassifier()

# 정확도
acc = cross_val_score(dt_clf, x_data, y_data, scoring = 'accuracy', cv=10)
print('교차 검증별 정확도: ', np.round(acc, 4))
# 정밀도
pre = cross_val_score(dt_clf, x_data, y_data, scoring = 'precision', cv=10)
print('교차 검증별 정밀도: ', np.round(pre,4))
# 재현율
rec = cross_val_score(dt_clf, x_data, y_data, scoring = 'recall', cv=10)
print('교차 검증별 재현율: ', np.round(rec, 4))
print('------------------------------------------------------------')
print('평균 - 정확도: {0:.4f}, 정밀도 : {1:.4f}, 재현율: {2:.4f}'
      .format(np.round(np.mean(acc),4),np.round(np.mean(pre),4),np.round(np.mean(rec),4)))
```

```python
kn_clf = KNeighborsClassifier(n_neighbors = 3)

# 정확도
acc = cross_val_score(kn_clf, x_data, y_data, scoring = 'accuracy', cv=10)
print('교차 검증별 정확도: ', np.round(acc, 4))
# 정밀도
pre = cross_val_score(kn_clf, x_data, y_data, scoring = 'precision', cv=10)
print('교차 검증별 정밀도: ', np.round(pre, 4))
# 재현율
rec = cross_val_score(kn_clf, x_data, y_data, scoring = 'recall', cv=10)
print('교차 검증별 재현율: ', np.round(rec, 4))
print('------------------------------------------------------------')
print('평균 - 정확도: {0:.4f}, 정밀도 : {1:.4f}, 재현율: {2:.4f}'
      .format(np.round(np.mean(acc),4),np.round(np.mean(pre),4),np.round(np.mean(rec),4)))
```

```python
lr_clf = LogisticRegression()

# 정확도
acc = cross_val_score(lr_clf, x_data, y_data, scoring = 'accuracy', cv=10)
print('교차 검증별 정확도: ', np.round(acc, 4))
# 정밀도
pre = cross_val_score(lr_clf, x_data, y_data, scoring = 'precision', cv=10)
print('교차 검증별 정밀도: ', np.round(pre, 4))
# 재현율
rec = cross_val_score(lr_clf, x_data, y_data, scoring = 'recall', cv=10)
print('교차 검증별 재현율: ', np.round(rec, 4))
print('------------------------------------------------------------')
print('평균 - 정확도: {0:.4f}, 정밀도 : {1:.4f}, 재현율: {2:.4f}'
      .format(np.round(np.mean(acc),4),np.round(np.mean(pre),4),np.round(np.mean(rec),4)))
```
